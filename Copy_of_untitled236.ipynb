```python
import pandas as pd
```

```python
# Load the Excel file
file_path = "/content/moAfrika (1).xlsx"  # Change this to your actual file path
xls = pd.ExcelFile(file_path)

# Display available sheet names
print(xls.sheet_names)
```

```python
# Load the "Tours" sheet into a DataFrame
df_tours = pd.read_excel(xls, sheet_name="Tours")

# Display the first few rows to understand the structure
df_tours.head()
```

```python
# Define keywords for classification
accommodation_keywords = {
    "Lodge": ["lodge"],
    "Camp": ["camp"],
    "Resort": ["resort"],
    "Hotel": ["hotel"],
    "Chalet": ["chalet"],
    "Tented Camp": ["tented camp", "tented"],
}

safari_keywords = {
    "Game Drive": ["game drive"],
    "Walking Safari": ["walking safari", "bush walk"],
    "Boat Safari": ["boat safari", "mokoro", "canoe"],
    "Fly-in Safari": ["fly-in", "air safari"],
}
```

```python
# Function to classify accommodationType
def classify_accommodation(text):
    if pd.isna(text):  # Handle missing values
        return None
    text = text.lower()  # Convert to lowercase for case-insensitive matching
    for category, keywords in accommodation_keywords.items():
        if any(keyword in text for keyword in keywords):
            return category
    return "Other"  # Default category if no match is found

# Function to classify safariType
def classify_safari(text):
    if pd.isna(text):  # Handle missing values
        return None
    text = text.lower()  # Convert to lowercase for case-insensitive matching
    for category, keywords in safari_keywords.items():
        if any(keyword in text for keyword in keywords):
            return category
    return "Other"  # Default category if no match is found
```

```python
# Apply classification functions
df_tours["accomodationType"] = df_tours["title"].apply(classify_accommodation)
df_tours["safariType"] = df_tours["description"].apply(classify_safari)

# Display updated dataframe with new columns
df_tours[["title", "accomodationType", "safariType"]].head()
```

```python
# Save to a new Excel file
df_tours.to_excel("moAfrika_updated.xlsx", index=False)

print("Updated dataset saved successfully!")
```

```python
def classify_accommodation(row):
    text = str(row["title"]) + " " + str(row["description"])  # Combine title and description
    text = text.lower()  # Convert to lowercase for consistent matching

    for category, keywords in accommodation_keywords.items():  # Loop through accommodation categories
        if any(keyword in text for keyword in keywords):  # Check if any keyword is found in text
            return category  # Return the matched category

    return "Other"  # Default if no match
```

```python
def classify_safari(row):
    text = str(row["title"]) + " " + str(row["description"])  # Combine title and description
    text = text.lower()  # Convert to lowercase for case-insensitive matching

    for category, keywords in safari_keywords.items():  # Loop through safari categories
        if any(keyword in text for keyword in keywords):  # Check for keyword match
            return category  # Return the matched category

    return "Other"  # Default if no match
```

```python
df_tours["accomodationType"] = df_tours.apply(classify_accommodation, axis=1)
df_tours["safariType"] = df_tours.apply(classify_safari, axis=1)
```
