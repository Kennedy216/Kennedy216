```python
confirmed = pd.read_csv("/time_series_covid_19_confirmed.csv")
deaths = pd.read_csv("/time_series_covid_19_deaths.csv")
recovered = pd.read_csv("/time_series_covid_19_recovered.csv")
```

```python
# Confirm data information
```

```python
confirmed.head()
```

```python
# Drop Unnecessary Columns
```

```python
confirmed = confirmed.drop(columns=["Province/State"])
deaths = deaths.drop(columns=["Province/State"])
recovered = recovered.drop(columns=["Province/State"])

```

```python
# Aggregate by Country
```

```python
confirmed = confirmed.groupby("Country/Region").sum().reset_index()
deaths = deaths.groupby("Country/Region").sum().reset_index()
recovered = recovered.groupby("Country/Region").sum().reset_index()
```

```python
# Convert Wide Format to Long Format
```

```python
confirmed_long = confirmed.melt(
    id_vars=["Country/Region"],
    var_name="Date",
    value_name="Confirmed"
)
```

```python
# Deaths by Country
```

```python
deaths_long = deaths.melt(
    id_vars=["Country/Region"],
    var_name="Date",
    value_name="Deaths"
)
```

```python
# Recovered Rate
```

```python
recovered_long = recovered.melt(
    id_vars=["Country/Region"],
    var_name="Date",
    value_name="Recovered"
)
```

```python
  # Datasets Merge
```

```python
df = confirmed_long.merge(
    deaths_long,
    on=["Country/Region", "Date"]
)

df = df.merge(
    recovered_long,
    on=["Country/Region", "Date"]
)
```

```python
# Convert all the Date Column
```

```python
print(confirmed_long["Date"].unique()[:20])
```

```python
confirmed_long = confirmed_long[
    ~confirmed_long["Date"].isin(["Lat", "Long"])
]
```

```python
print(confirmed_long["Date"].unique()[:20])
```

```python
# Convert Date Column
```

```python
confirmed_long["Date"] = pd.to_datetime(
    confirmed_long["Date"],
    format="%m/%d/%y"
)
```

```python
# Repeat Cleaning for Deaths and Recovered
```

```python
deaths_long = deaths_long[
    ~deaths_long["Date"].isin(["Lat", "Long"])
]

recovered_long = recovered_long[
    ~recovered_long["Date"].isin(["Lat", "Long"])
]
```

```python
# Convert their Dates
```

```python
deaths_long["Date"] = pd.to_datetime(
    deaths_long["Date"],
    format="%m/%d/%y"
)

recovered_long["Date"] = pd.to_datetime(
    recovered_long["Date"],
    format="%m/%d/%y"
)
```

```python
# Merge all Three Datasets
```

```python
df = confirmed_long.merge(
    deaths_long,
    on=["Country/Region", "Date"]
)

df = df.merge(
    recovered_long,
    on=["Country/Region", "Date"]
)
```

```python
# Rename Column
```

```python
df = df.rename(columns={"Country/Region": "Country"})
```

```python
# Sort the Data (Important for Time Series)
```

```python
df = df.sort_values(["Country", "Date"])
```

```python
# Create Basic Metrics
```

```python
df["Mortality_Rate"] = df["Deaths"] / df["Confirmed"]
```

```python
df["Recovery_Rate"] = df["Recovered"] / df["Confirmed"]
```

```python
df["Daily_New_Cases"] = df.groupby("Country")["Confirmed"].diff()
```

```python
df = df.fillna(0)
```

```python
# Export to Excel
```

```python
df.to_excel("covid_cleaned.xlsx", index=False)
```
